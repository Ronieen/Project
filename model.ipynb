```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
athlet = open('/mnt/HC_Volume_18315164/home-jupyter/jupyter-l-lev/athlete_events.csv')
#Открытие файла
```

```python
container = []
for row in athlet:
    container.append(row.split(','))
data = pd.DataFrame(container)
#Функция, которая делит строки по запятой
```

```python
data = data.applymap(lambda x: x.replace('\n', '').replace(';', '').replace('""', '').replace('"', '') if x is not None else x)
#Приведение дааных к приемлемому виду
```

```python
data.columns = data.iloc[0]
data = data[1:]
```

```python
data = data.drop(data.columns[[15, 16, 17, 18]], axis=1)
```

```python
data = data.drop_duplicates()
```

```python
data
```

```python
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Height'] = pd.to_numeric(data['Height'], errors='coerce')
data['Weight'] = pd.to_numeric(data['Weight'], errors='coerce')
#Все недопустимые символы меняю на пустое значение
```

```python
atl_df = data[(data['Event'] == "Athletics Men's 100 metres")]
#Делаем выборку
```

```python
atl_df['Medal'] = atl_df['Medal'].replace({'Gold': 1, 'Silver': 2, 'Bronze': 3, None: 0, 'NA': 0})
#Заменяем значения в столбце медали на 1 и 0
```

```python
exel_df = atl_df[(atl_df['Medal'] == 1)]
#Берем тольок тех, кто заработал золотые медали(1)
```

```python
exel_df = exel_df.loc[:, ["Age", "Height", 'Weight', "Year", "Medal", "Name"]]
#Отбираем только нужны столбцы для дальнейшей работе
```

```python
exel_df = exel_df.sort_values(by='Year', ascending=True)
#Сортировка по годам
```

```python
exel_df.to_excel('файл.xlsx', index=False)
```

```python
exel_df = exel_df.drop(exel_df.index[-1], axis=0)
```

```python
exel_df
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.utils import resample
import numpy as np

# Объединение дубликатов по годам с помощью среднего значения
exel_df = exel_df.groupby('Year').mean().reset_index()

# Интерполяция отсутствующих значений
exel_df.interpolate(method='linear', inplace=True)

# Выбор признаков и целевой переменной
X = exel_df[['Year']]
y = exel_df['Age']

# Создание и обучение модели линейной регрессии
model = LinearRegression()
model.fit(X, y)

# Предсказание для 2016 года
prediction = model.predict([[2016]])

# Расчет доверительного интервала
predictions = []
for _ in range(1000):  
    sample_X, sample_y = resample(X, y)
    model.fit(sample_X, sample_y)
    predictions.append(model.predict([[2016]]))

confidence_interval = np.percentile(predictions, [2.5, 97.5])  # Доверительный интервал 95%

print(f"Предсказание возраста в 2016 году: {prediction[0]}")
print(f"95% доверительный интервал: {confidence_interval}")
```
